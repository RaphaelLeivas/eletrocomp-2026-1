```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
omega = 2 # Frequency
t_max = 10
t = np.linspace(0, t_max, 1000) # Time vector
signal0 = np.exp(0 + 1j * omega * t) 
signal1 = np.exp(0.1 + 1j * omega * t) 
signal2 = np.exp(-0.1 + 1j * omega * t) 

# Plot 3D Helix
fig = plt.figure(figsize=(12, 6))
ax = fig.add_subplot(111, projection='3d')

ax.plot(t, np.real(signal0), np.imag(signal0), color="red", label=r'$\alpha = 0$')
ax.plot(t, np.real(signal1), np.imag(signal1), color="blue", label=r'$\alpha = 0.1$')
ax.plot(t, np.real(signal2), np.imag(signal2), color="green", label=r'$\alpha = -0.1$')
ax.set_zlabel('Z axis', labelpad=0)
ax.set_xlabel('Tempo (t)')
ax.set_ylabel('Real')
ax.set_zlabel('Imaginário')

plt.title(r'Solução $e^{(\alpha + j\omega)t} \, , \, \omega = 2$')
plt.legend()
plt.tight_layout()
plt.subplots_adjust(right=0.95)
plt.show()
```

```python
omega = 2
y_0 = 1
delta_t = 0.1
number_of_steps = int(t_max / delta_t)
y_n_0 = [y_0]
y_n_1 = [y_0]
y_n_2 = [y_0]

for i in range(1, number_of_steps):
    y_n_0.append(y_n_0[i-1] + delta_t * (0 + omega * 1j * y_n_0[i-1]) ) 
    y_n_1.append(y_n_1[i-1] + delta_t * (0.1 + omega * 1j * y_n_1[i-1]) )
    y_n_2.append(y_n_2[i-1] + delta_t * (-0.1 + omega * 1j * y_n_2[i-1]) )

t = np.linspace(0, t_max, number_of_steps)

fig = plt.figure(figsize=(12, 6))
ax = fig.add_subplot(111, projection='3d')

ax.plot(t, np.real(y_n_0), np.imag(y_n_0), color="red", label=r'$\alpha = 0$')
ax.plot(t, np.real(y_n_1), np.imag(y_n_1), color="blue", label=r'$\alpha = 0.1$')
ax.plot(t, np.real(y_n_2), np.imag(y_n_2), color="green", label=r'$\alpha = -0.1$')
ax.set_zlabel('Z axis', labelpad=-5)
ax.set_xlabel('Tempo (t)')
ax.set_ylabel('Real')
ax.set_zlabel('Imaginário')

plt.title("Solução numérica via Método de Euler")
plt.legend()
plt.legend()
plt.tight_layout()
plt.subplots_adjust(right=0.95)
plt.show()
```
