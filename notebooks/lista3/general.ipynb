```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq

# --- parâmetros físicos ---
eps = 8.854187817e-12  # Vacuum permittivity (F/m)
mu = 4 * np.pi * 1e-7  # Vacuum permeability (H/m)
vp = 1 / np.sqrt(eps * mu)

# --- domínio ---
dt = 50e-12
T = 100 * dt
Nt = int(T / dt)

t_grid = np.linspace(0, Nt, Nt)
source_gaussiano = np.exp(-((t_grid * dt - 1.8e-9)**2) / (0.6e-9)**2)

# Create a figure with 1 row and 2 columns
fig, ax = plt.subplots(1, 1, figsize=(8, 5))

ax.plot(t_grid * dt, source_gaussiano)
ax.set_title("Pulso Gaussiano")
ax.set_xlabel("t (ns)")
ax.set_ylabel("Amplitude")

plt.tight_layout() # Adjusts spacing between subplots
plt.show()
```

```python
# aplica a FFT do pulso gaussiano
# 2. Compute the FFT
yf = fft(source_gaussiano)
xf = fftfreq(len(source_gaussiano), dt)

# 3. Plot the results (Positive frequencies only)
plt.plot(xf[:len(source_gaussiano)//8], 1e3 * 2.0/len(source_gaussiano) * np.abs(yf[:len(source_gaussiano)//8]))
plt.title("Espectro FFT do pulso Gaussiano")
plt.xlabel(r"Frequência $\cdot 10^{9}$ Hz")
plt.ylabel(r"Amplitude $\cdot 10^{-3}$")
plt.grid()
plt.show()
```
